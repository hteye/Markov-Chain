```python
import numpy as np
import random as rm

# The statespace
states = ["Rest","Sweeties","Exercise"]

# Possible sequences of events
transitionName = [["RR","RE","RS"],["ER","EE","ES"],["SR","SE","SS"]]

# Probabilities matrix (transition matrix)
transitionMatrix = [[0.2,0.6,0.2],[0.1,0.6,0.3],[0.2,0.7,0.1]]

def predictor(days):
    # Choose the starting state
    activityToday = "Rest"
    activityList = [activityToday]
    i = 0
    prob = 1
    while i != days:
        if activityToday == "Rest":
            change = np.random.choice(transitionName[0],replace=True,p=transitionMatrix[0])
            if change == "RR":
                prob = prob * 0.2
                activityList.append("Rest")
                pass
            elif change == "RE":
                prob = prob * 0.6
                activityToday = "Exercise"
                activityList.append("Exercise")
            else:
                prob = prob * 0.2
                activityToday = "Sweeties"
                activityList.append("Sweeties")
        elif activityToday == "Exercise":
            change = np.random.choice(transitionName[1],replace=True,p=transitionMatrix[1])
            if change == "EE":
                prob = prob * 0.5
                activityList.append("Exercise")
                pass
            elif change == "ER":
                prob = prob * 0.2
                activityToday = "Rest"
                activityList.append("Rest")
            else:
                prob = prob * 0.3
                activityToday = "Sweeties"
                activityList.append("Sweeties")
        elif activityToday == "Sweeties":
            change = np.random.choice(transitionName[2],replace=True,p=transitionMatrix[2])
            if change == "SS":
                prob = prob * 0.1
                activityList.append("Sweeties")
                pass
            elif change == "SR":
                prob = prob * 0.2
                activityToday = "Rest"
                activityList.append("Rest")
            else:
                prob = prob * 0.7
                activityToday = "Exercise"
                activityList.append("Exercise")
        i += 1    
    return activityList

# To save every activityList
list_activity = []
count_sweeties = 0
count_exercise = 0
count_rest = 0

#Run the experiment 1000 times to find expected probabilities
for i in range(1,1000):
        list_activity.append(predictor(15))

# Iterate through the list to get a count of all activities ending in state:'Run'
for lst in list_activity:
    if(lst[2] == "Sweeties"):
        count_sweeties += 1
    elif(lst[2] == "Exercise"):
        count_exercise += 1
    if(lst[2] == "Rest"):
        count_rest += 1

# Calculate the probability of starting from state:'Rest' and ending at state 'Rest' after 15 days
percentage = (count_rest/1000) * 100
print("The probability of starting at state:'Rest' and ending at state:'Rest' after 15 days = " + str(percentage) + "%")

# Calculate the probability of starting from state:'Rest' and ending at state 'Exercise' after 15 days
percentage = (count_exercise/1000) * 100
print("The probability of starting at state:'Rest' and ending at state:'Exercise' after 15 days = " + str(percentage) + "%")

# Calculate the probability of starting from state:'Rest' and ending at state 'Sweeties' after 15 days
percentage = (count_sweeties/1000) * 100
print("The probability of starting at state:'Rest' and ending at state:'Sweeties' after 15 days = " + str(percentage) + "%")
print("\n")

list_activity = []
count_sweeties = 0
count_exercise = 0
count_rest = 0
#Run the experiment 1000 times to find expected probabilities
for i in range(1,1000):
        list_activity.append(predictor(10000))

# Iterate through the list to get a count of all activities ending in state:'Run'
for lst in list_activity:
    if(lst[2] == "Sweeties"):
        count_sweeties += 1
    elif(lst[2] == "Exercise"):
        count_exercise += 1
    if(lst[2] == "Rest"):
        count_rest += 1

# Calculate the probability of starting from state:'Rest' and ending at state 'Rest' after 15 days
percentage = (count_rest/1000) * 100
print("The probability of starting at state:'Rest' and ending at state:'Rest' after 10000 days = " + str(percentage) + "%")

# Calculate the probability of starting from state:'Rest' and ending at state 'Exercise' after 15 days
percentage = (count_exercise/1000) * 100
print("The probability of starting at state:'Rest' and ending at state:'Exercise' after 10000 days = " + str(percentage) + "%")

# Calculate the probability of starting from state:'Rest' and ending at state 'Sweeties' after 15 days
percentage = (count_sweeties/1000) * 100
print("The probability of starting at state:'Rest' and ending at state:'Sweeties' after 10000 days = " + str(percentage) + "%")
```
